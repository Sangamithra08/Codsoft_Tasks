```python
#implementinh undersampling to handle imbalaced dataset
from imblearn.under_sampling import NearMiss

# Assuming x and y are your feature and target variables
nm = NearMiss()
x_1, y_1 = nm.fit_resample(x, y)

print(x_1.shape)
print(y_1.shape)
```

```python
x_1.head()
y_1.head()
```

```python
plt.figure(figsize=(8,6))
plt.figure(figsize=(8, 6))
sns.countplot(x=y_1, hue=y_1)
plt.show()
```

DataSet is now balanced

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test=train_test_split(x_1,y_1,test_size=0.20,random_state=40)
```

```python
model = LogisticRegression(max_iter=1000)
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)

# Apply logistic regression on the scaled data
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
model.score(x_test,y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_predicted)
print(cm)
```
